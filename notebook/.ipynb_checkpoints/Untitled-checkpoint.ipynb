```python
import matplotlib.pyplot as plt
import pandas as pd
from config import (api_key, gkey)
import gmaps
```

```python
brooklyn_housing_data = "../raw_data/brooklyn_sales_map.csv"
housing_data = pd.read_csv(brooklyn_housing_data)
housing_data_df = pd.DataFrame(housing_data)
housing_data_df_clean = housing_data_df.rename(columns={'year_of_sale': 'Years',
                                                       'neighborhood': 'Neighborhoods'})
housing_data_df_clean.head()
```

```python
total_neighborhoods = housing_data_df_clean.groupby(['Years',
                                            'Neighborhoods'])
#total_neighborhoods.head()

number_of_sales = total_neighborhoods['sale_price'].count()
#number_of_sales

avg_sale_price = total_neighborhoods['sale_price'].mean()
#avg_sale_price

neighborhood_data = {'Total Number of Sales by Neighborhood': number_of_sales,
                       'Average Sale Price by Neighborhood': avg_sale_price}
#neighborhood_data

neighborhood_summary = pd.DataFrame(neighborhood_data)
#neighborhood_summary

final_neighborhood_summary = neighborhood_summary.copy()
final_neighborhood_summary['Average Sale Price by Neighborhood'] = final_neighborhood_summary['Average Sale Price by Neighborhood'].map('${:,.0f}'.format)
final_neighborhood_summary.head()
```

```python
final_neighborhood_summary.unstack()
final_neighborhood_summary.unstack(level=0).head()
final_neighborhood_summary.unstack(level=0).columns
```

```python
x_axis = [2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
final_neighborhood_summary.unstack(level=0).plot(kind='scatter', x = x_axis, y='Total Number of Sales by Neighborhood')
plt.show
```

```python
gmaps.configure(api_key=gkey)

fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(housing_data_df_clean['Neighborhoods'], weights= housing_data_df_clean['sale_price'], 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 1)

heat_layer.dissipating = False
heat_layer.max_intensity = 100
heat_layer.point_radius = 1

fig.add_layer(heat_layer)

fig
```
