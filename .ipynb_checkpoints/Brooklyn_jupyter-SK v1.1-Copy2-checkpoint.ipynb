```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
file = "brooklyn_sales_map.csv"
file_df = pd.read_csv(file)
file_df.head()
```

```python
file_df.columns[0:20]
```

```python
bk_master = file_df[['Unnamed: 0', 'building_class_category','sale_price','gross_sqft',
                     'year_of_sale','year_built', "ZipCode", 'neighborhood','residential_units', 'SchoolDist',
                     'BldgClass', 'YearAlter1', 'YearAlter2', 'LotArea','ResArea', 'UnitsRes', 'NumFloors',  
                    ]]
bk_master.head()
```

```python
# Master csv filtered for one family homes
one_fam_master = bk_master.loc[bk_master["building_class_category"] == "01 ONE FAMILY HOMES",:] 
print(len(one_fam_master))
one_fam_master.head()
```

```python
one_fam_master.dtypes
```

```python
#One Family Master filtered for three years data
one_fam_3years = one_fam_3years.dropna(how='any')
one_fam_3years = one_fam_master.loc[(one_fam_master["year_of_sale"] == 2015  )|
                                    (one_fam_master["year_of_sale"] == 2016  )|
                                    (one_fam_master["year_of_sale"] == 2017  ) ,:]

one_fam_3years = one_fam_3years.loc[(one_fam_3years["sale_price"]<5000000)&
                                    (one_fam_3years["sale_price"] > 100000)& 
                                    (one_fam_3years["gross_sqft"] < 5000)& 
                                    (one_fam_3years["gross_sqft"] > 100)& 
                                    (one_fam_3years["year_built"] > 1890)
                                    ,:]
one_fam_3years["SP/Sq_ft"] = one_fam_3years["sale_price"] / one_fam_3years["gross_sqft"]
print(len(one_fam_3years))
one_fam_3years.head()
```

```python
# Sale Price vs Area
N = len(one_fam_3years)
x = one_fam_3years['gross_sqft']
y = one_fam_3years['sale_price']
colors = ("blue")
area = np.pi*15

# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Sale Price vs Area (sq-ft)')
plt.xlabel('gross_sqft')
plt.ylabel('Sale Price')
plt.show()
```

```python
# Sale Price vs Lot Area
N = len(one_fam_3years)
x = one_fam_3years['LotArea']
y = one_fam_3years['sale_price']
colors = ("green")
area = np.pi*15

# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Sale Price vs Lot Area')
plt.xlabel('Lot Area')
plt.ylabel('Sale Price')
plt.show()
```

```python
# Sale Price vs Year Built
N = len(one_fam_3years)
x = one_fam_3years['year_built']
y = one_fam_3years['sale_price']
colors = ("red")
area = np.pi*15

# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Sale Price vs Year Built')
plt.xlabel('Year Built')
plt.ylabel('Sale Price')
plt.show()

```

```python
# Sale Price vs Year Built
N = len(one_fam_3years)
x = one_fam_3years['SchoolDist']
y = one_fam_3years['sale_price']
colors = ("black")
area = np.pi*15

# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Sale Price vs School District')
plt.xlabel('School District')
plt.ylabel('Sale Price')
plt.show()
```

```python
#SP histogram
sns.distplot(one_fam_3years['sale_price']);
#skewness and kurtosis
print("Skewness: %f" % one_fam_3years['sale_price'].skew())
print("Kurtosis: %f" % one_fam_3years['sale_price'].kurt())
```

### Characteristics: 
    1. Deviates from the normal distribution.
    2. Has appreciable positive skewness.
    3. Demonstrates peakedness.

```python
#box plot zipcode/saleprice
var = one_fam_3years["ZipCode"]
SP = one_fam_3years["sale_price"]
data = pd.concat([SP, var], axis=1)
f, ax = plt.subplots(figsize=(20, 6))
fig = sns.boxplot(x=var, y=SP, data=data)
fig.axis(ymin=0, ymax=5000000);
```

```python
#box plot zipcode vs SP/sq.ft
var = one_fam_3years["ZipCode"]
SF = one_fam_3years["SP/Sq_ft"]
data = pd.concat([SF, var], axis=1)
f, ax = plt.subplots(figsize=(20, 6))
fig = sns.boxplot(x=var, y=SP, data=data)
fig.axis(ymin=0, ymax=3000);
```

```python
#box plot Neighbourhood vs SP/sq.ft
var = one_fam_3years["neighborhood"]
SP = one_fam_3years["SP/Sq_ft"]
data = pd.concat([SP, var], axis=1)
f, ax = plt.subplots(figsize=(20, 6))
fig = sns.boxplot(x=var, y=SP, data=data)
fig.axis(ymin=0, ymax=3000);
```

```python
#box plot School Dist vs SP/sq.ft
var = one_fam_3years["SchoolDist"]
SP = one_fam_3years["SP/Sq_ft"]
data = pd.concat([SP, var], axis=1)
f, ax = plt.subplots(figsize=(20, 6))
fig = sns.boxplot(x=var, y=SP, data=data)
fig.axis(ymin=0, ymax=3000);
```

```python
#correlation matrix
corrmat = one_fam_3years.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
#saleprice correlation matrix
k = 15 #number of variables for heatmap
cols = corrmat.nlargest(k, 'sale_price')['sale_price'].index
cm = np.corrcoef(one_fam_3years[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 5}, 
                 yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
#applying log transformation
one_fam_3years['sale_price'] = np.log(one_fam_3years['sale_price'])

#transformed histogram and normal probability plot
sns.distplot(one_fam_3years['sale_price'], fit=norm);
fig = plt.figure()
res = stats.probplot(one_fam_3years['sale_price'], plot=plt)
```
