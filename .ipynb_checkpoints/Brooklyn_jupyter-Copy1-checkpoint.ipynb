```python
%matplotlib notebook
import random
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = "brooklyn_sales_map.csv"
file_df = pd.read_csv(file)
file_df.head()
```

```python
file_df.columns
```

```python
building_class_price = file_df[['building_class_category','sale_price','year_built']]
building_class_price.head()
```

```python
building_class_price = building_class_price[building_class_price['year_built'] > 1700]
```

```python
x_axis = building_class_price['year_built']
y_axis = building_class_price['sale_price']
plt.xlabel('Year Built')
plt.ylabel('Sale Price')
plt.title('Correlation between building year built and sale price')
plt.scatter(x_axis,y_axis,marker = 'o',facecolor = 'red',edgecolor = 'black',alpha = 0.75)
```
