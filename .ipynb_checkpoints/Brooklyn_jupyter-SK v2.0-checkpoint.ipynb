```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import sys
sys.path.append('Desktop/Columbia%20Engineering/Project%20Uno/SK%20Project%20Folder/')
import exp
```

```python
file = "brooklyn_sales_map.csv"
file_df = pd.read_csv(file)
file_df.head()
```

```python
file_df.columns[0:20]
```

```python
bk_master = file_df[['Unnamed: 0', 'building_class_category','sale_price','gross_sqft',
                     'year_of_sale','year_built', "ZipCode", 'neighborhood','residential_units', 'SchoolDist',
                     'BldgClass', 'YearAlter1', 'YearAlter2', 'LotArea','ResArea', 'UnitsRes', 'NumFloors',  
                    ]]
bk_master.head()
```

```python
# Master csv filtered for one family homes
one_fam_master = bk_master.loc[bk_master["building_class_category"] == "01 ONE FAMILY HOMES",:] 
print(len(one_fam_master))
one_fam_master.head()
```

```python
#One Family Master filtered for three years data

one_fam_3years = one_fam_master.loc[(one_fam_master["year_of_sale"] == 2015  )|
                                    (one_fam_master["year_of_sale"] == 2016  )|
                                    (one_fam_master["year_of_sale"] == 2017  ) ,:]

one_fam_3years = one_fam_3years.loc[(one_fam_3years["sale_price"]<3000000)&
                                    (one_fam_3years["sale_price"] > 100000)& 
                                    (one_fam_3years["gross_sqft"] < 3000)& 
                                    (one_fam_3years["gross_sqft"] > 500)& 
                                    (one_fam_3years["year_built"] > 1890)
                                    ,:]
one_fam_3years["SP/Sq_ft"] = one_fam_3years["sale_price"] / one_fam_3years["gross_sqft"]

one_fam_3years = one_fam_3years.dropna(how='any')
one_fam_3years = one_fam_3years[one_fam_3years.SchoolDist != "NA"]

print(len(one_fam_3years))
one_fam_3years.head()
```

```python
#Add Alphabetical School District for Plot
one_fam_3years["SchoolDist_A"] = one_fam_3years["SchoolDist"] 
one_fam_3years["SchoolDist_A"]= one_fam_3years["SchoolDist_A"].apply(str)
one_fam_3years["SchoolDist_A"] = one_fam_3years["SchoolDist_A"].replace(
    {'13.0': 'S13', '14.0': 'S14','15.0': 'S15','16.0': 'S16','17.0': 'S17','18.0': 'S18','19.0': 'S19','20.0': 'S20',
     '21.0': 'S21', '22.0': 'S22','23.0': 'S23','32.0': 'S32'})
```

```python
# Sale Price vs Area
N = len(one_fam_3years)
x = one_fam_3years['gross_sqft']
y = one_fam_3years['sale_price']
colors = ("blue")
area = np.pi*15

# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Sale Price vs Area (sq-ft)')
plt.xlabel('gross_sqft')
plt.ylabel('Sale Price')
plt.show()
```
