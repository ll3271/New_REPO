```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = "brooklyn_sales_map.csv"
file_df = pd.read_csv(file)
file_df.head()
```

```python
building_area = file_df[['building_class_category','sale_price','gross_sqft']]
building_area
```

```python
x_axis = building_area['gross_sqft']
y_axis = building_area['sale_price']
plt.xlabel('Sq Ft')
plt.ylabel('Sale Price')
plt.title('Correlation between building size and sale price')
plt.scatter(x_axis,y_axis,marker = 'o',facecolor = 'red',edgecolor = 'black',alpha = 0.75)
```
